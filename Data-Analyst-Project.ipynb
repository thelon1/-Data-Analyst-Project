```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# Exploratary Data Analysis

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender' , data = df )

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = False)
```

```python
sns.barplot(x = 'Gender', y = 'Amount' , data = sales_gen)
plt.show()
```

```python
# From above Graph most  of the buyers are female  and even purchasing power of female are greater than men.
```

# Age.

```python
df.columns
```

```python
ax = sns.countplot(x = 'Age Group', data = df , hue = 'Gender')

for bar in ax.containers:
    ax.bar_label(bar)
    
plt.show()
```

```python
sale_age = df.groupby(['Age Group'] , as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = False)
```

```python
sns.barplot(x = 'Age Group', y = 'Amount', data =  sale_age, )

plt.show()
```

#### From the above graph we can see that most of the buyers are of age group 23yrs - 35yrs female.

# State

```python
df.columns
```

```python
sale_state = df.groupby(['State'] , as_index = False)['Orders'].sum().sort_values(by = 'Orders' , ascending = False).head(9)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x = 'State' , y = 'Orders' , data = sale_state )

plt.show()
```

```python
sale_state = df.groupby(['State'] , as_index = False)['Amount'].sum().sort_values(by='Amount' ,ascending = False ).head(9)

sns.set(rc={'figure.figsize':(15,5)})

sns.barplot(data = sale_state , x = 'State' , y = 'Amount')

plt.show()
```

#### From above graph most of the order & total/sales&amount are from utarpradesh, maharashtra , karnataka.

# Marital_Status

```python
ax = sns.countplot(data = df , x = 'Marital_Status')

sns.set(rc={'figure.figsize':(6,5)})

for bar in ax.containers:
    ax.bar_label(bar)
    
```

```python
custom_palette = ['#FF5733', '#33FF57', '#3357FF']
sns.set_palette(custom_palette)
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender',palette=["#fc9272", "#fee0d1"])
```

#### From the above graph  we can see that most of the buyes are married (women)  and they have high purchasing power.

```python
ax = sns.countplot(data = df , x = 'Occupation')
sns.set(rc={'figure.figsize':(26,5)})
for bar in ax.containers:
    ax.bar_label(bar)
    
```

```python
sale_state = df.groupby(df['Occupation'] , as_index =False)['Amount'].sum().sort_values(by='Amount' ,ascending = False)
```

```python
sns.set(rc={'figure.figsize':(22,5)})
```

```python

sns.barplot(data = sale_state , x ='Occupation', y ='Amount')
plt.show()
```

#### From the above graph we can see that most of the buyers are  wroking in IT Sector , Healthcare , Aviation sector

# Product Category

```python
sns.set(rc={'figure.figsize':(26,5)})
ax = sns.countplot(data = df , x = 'Product_Category')
for bar in ax.containers:
    ax.bar_label(bar)

plt.show()
```

```python
sales_state = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending = False).head(10)
```

```python
    sns.set(rc={'figure.figsize':(27,5)})
    sns.barplot(data = sales_state , x = 'Product_Category' , y='Amount' )
    plt.show()
```

#### From the above graph we can see that most of the sold product are from food, Clthing & Approval ,and electronic category

```python
sales_state = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by='Orders',ascending = False).head(10)
sns.set(rc={'figure.figsize':(27,5)})
sns.barplot(data = sales_state , x = 'Product_ID' , y='Orders' )
plt.show()
```

* Married women age group  26-35yrs from Up, Maharashtra and karnataka in IT,Healthcare and Avition are more likely to buy product from food, clothing and electronics Category
